```python
import pandas as pd
```

```python
def overs_to_balls(over):
    whole = int(over)
    balls = round((over - whole)*10)
    if balls > 5:
        raise ValueError("Invalid Over Format! Balls should be between 0 to 5")
    return whole*6 + balls
```

```python
class Team:
    def __init__(self,name):
        self.name = name
        self.matches = 0
        self.wins = 0
        self.losses = 0
        self.points = 0
        self.runs_scored = 0
        self.balls_faced = 0
        self.runs_conceded = 0
        self.balls_bowled = 0
    def update_stats(self,runs_scored,overs_faced,wickets_lost,runs_conceded,overs_bowled,wickets_taken):
        self.matches += 1
        #batting overs
        if wickets_lost == 10:
            self.balls_faced += 120
        else:
            self.balls_faced += overs_to_balls(overs_faced)
    #bowling overs
        if wickets_taken == 10:
            self.balls_bowled += 120
        else:
            self.balls_bowled += overs_to_balls(overs_bowled)
        self.runs_scored += runs_scored
        self.runs_conceded += runs_conceded
        if runs_scored > runs_conceded:
            self.wins += 1
            self.points += 2
        else:
            self.losses += 1
    def calculate_nrr(self):
        if self.balls_faced == 0 or self.balls_bowled == 0:
            return 0
        else:
            run_rate_scored = (self.runs_scored/self.balls_faced)*6
            run_rate_conceded = (self.runs_conceded/self.balls_bowled)*6
            return round(run_rate_scored - run_rate_conceded, 3)
```

```python
teams = {}
num_matches = int(input("Enter Number of matches: "))
for _ in range(num_matches):
    home_team = input("\n Enter Home Team name: ")
    away_team = input("Enter Away Team name: ")
    if home_team not in teams:
        teams[home_team] = Team(home_team)
    if away_team not in teams:
        teams[away_team] = Team(away_team)
    print(f"\n------{home_team} Innings ------")
    home_team_runs = int(input("Enter Number of runs scored: "))
    home_team_wickets = int(input("Enter Number of wickets lost: "))
    home_team_overs = float(input("Enter Number of overs faced: "))
    
    print(f"\n------{away_team} Innings ------")
    away_team_runs = int(input("Enter Number of runs scored: "))
    away_team_wickets = int(input("Enter Number of wickets lost: "))
    away_team_overs = float(input("Enter Number of overs faced: "))
    
    teams[home_team].update_stats(home_team_runs,home_team_overs,home_team_wickets,away_team_runs,away_team_overs,away_team_wickets)
    teams[away_team].update_stats(away_team_runs,away_team_overs,away_team_wickets,home_team_runs,home_team_overs,home_team_wickets)
    print("\nScores Taken Successfully")
```

```python
#Pandas Structure

data = []
for team in teams.values():
    data.append({
        "Team": team.name,
        "Matches": team.matches,
        "Wins": team.wins,
        "Losses": team.losses,
        "Points": team.points,
        "NRR": team.calculate_nrr()
    })
df = pd.DataFrame(data)

df = df.sort_values(by=["Points","NRR"], ascending = False)

print("\n POINTS TABLE")
print(df.to_string(index = False))
df.to_csv("points_table.csv", index = False)
```
